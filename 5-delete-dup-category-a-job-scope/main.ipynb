```python
import pandas as pd
```

```python
# read excel
df = pd.read_excel('data.xlsx')
df
```

```python
# get data that will not be included in the deduplication
category_to_delete = 'A'
filt_preserved = df['CATEGORY'] != category_to_delete
df_preserved = df[filt_preserved]
df_preserved
```

```python
# get data that will be included in the deduplication
filt_delete = df['CATEGORY'] == category_to_delete
df_to_dedup = df[filt_delete]
df_to_dedup
```

```python
# remove duplicates
df_deduped = df_to_dedup.drop_duplicates(subset=['JOB SCOPE','DATE'])
df_deduped
```

```python
# combine the preserved and deduplicated data
final_df = pd.concat([df_preserved,df_deduped])
# sort data
final_df_sorted = final_df.sort_values(by=["CATEGORY"])
final_df_sorted
```

```python
# format dates
final_df_sorted["DATE"] = pd.to_datetime(final_df_sorted["DATE"]).dt.date

# output to excel
final_df_sorted.to_excel("output.xlsx", index=False)
```
