```python
# import dependencies
import pandas as pd
```

```python
# read excel file
df0 = pd.read_excel('data0.xlsx') # old data
df0
```

```python
# read excel file
df1 = pd.read_excel('data1.xlsx') # new data
df1
```

```python
# merge the two files
merged_df = pd.merge(df0,df1,how='right')

# format dates
merged_df['Entry Date'] = pd.to_datetime(merged_df['Entry Date']).dt.date
merged_df['Received Date'] = pd.to_datetime(merged_df['Received Date']).dt.date
merged_df
```

```python
# output to excel
merged_df.to_excel('output.xlsx',index=False)
```
